```python
class student:
    def __init__(self, name, score):
        self.name=name
        self.score=score
    def display(self):
        print("name: "+self.name)
        print("score: ",self.score)
s1=student("Tom", 80)
s2=student("Mary", 90)
s1.display()
s2.display()
        
```

```python
class item:
    def __init__(self):
        self.name = input("Enter name:")
        self.quantity = int(input("Enter quantity:"))
        self.cost = int(input("Enter cost:"))
        
    def display(self):
        print("name: ", self.name)
        print("quantity ", self.quantity)
        print("cost: ", self.cost)
        
    def getTotal(self):
        return self.quantity * self.cost
s1 = item()
s2 = item()
s3 = item()
total = s1.getTotal() + s2.getTotal() + s3.getTotal()
print("The total is: ", total)
        
    
```

```python
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
class student(person):
    def __init__(self, name, age, score):
        person.__init__(self, name, age)
        self.score = score
    def display(self):
        print("name: "+self.name)
        print("age: ",self.age)
        print("score: ",self.score)
        
s1=student("Dio", 20, 80)
s2=student("Jotaro", 19, 79)
s1.display()
s2.display()
    
```
