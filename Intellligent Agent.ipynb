# Simple Reflex Agent

```python
import random
class environment(object): #class for environment
    def __init__(self):
        #instantiate locations and conditions
        #0 indicates clean, 1 indicates dirty
        self.location = ["a", "b"]
        self.locationcondition = {"a":"0", "b":"0"}
        
        #randomize the conditions
        self.locationcondition ["a"] = random.randint (0, 1) #random.randint (0, 1)
        self.locationcondition ["b"] = random.randint (0, 1)
        
        #place vacuum at random location
        self.vacuumlocation = random.choice(self.location) #random.choice

class agent (environment): # class for agent - cleaning based on the current state of the environment - fully observable environment
    def __init__(self, environment):
        #randomize conditions
        print("location condition", environment.locationcondition)
        print("vacuum location", environment.vacuumlocation)
        count = 0
        while count < 2: # two rooms
            if (environment.locationcondition [environment.vacuumlocation] == 1): 
                environment.locationcondition [environment.vacuumlocation] = 0
                print(environment.vacuumlocation, "has been cleaned")
            else:
                print(environment.vacuumlocation, "is already clean")
            newindex = environment.location.index (environment.vacuumlocation) + 1
            if (newindex == 2):
                newindex = 0 # resets
            environment.vacuumlocation = environment.location [newindex]
            count = count + 1 # go to the next room
        print(environment.locationcondition)

e1 = environment() # objects
a1 = agent(e1)
```
