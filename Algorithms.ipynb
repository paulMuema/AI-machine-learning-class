```python
weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 
           'Overcast', 'Overcast', 'Rainy'] #list
play = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']

from sklearn import preprocessing
import numpy as np
from sklearn.naive_bayes import GaussianNB

lcode = preprocessing.LabelEncoder()
encoded_weather = lcode.fit_transform(weather) #encoded weather
encoded_play = lcode.fit_transform(play)
print(encoded_weather)

weather2d = np.reshape(encoded_weather, (-1, 1))
print(weather2d)
model = GaussianNB() #constructor
model.fit(weather2d, encoded_play)

p = model.predict([[2]])
print(p)
#it works as a classifier
```
