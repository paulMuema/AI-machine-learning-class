```python
import constraint
problem = constraint.Problem()
problem.addVariable("x", [1, 2, 3]) #adds variables and domains
problem.addVariable("y", range(10))

#def used every time when creating a function
def myConstraint(x, y):
    if x + y >= 5: #creating the rule
        return True
    
problem.addConstraint(myConstraint, ["x", "y"])
solutions = problem.getSolutions()
for s in solutions:
    print(s)
```

```python
import constraint
problem = constraint.Problem()

#create variable and domain
problem.addVariables("TF", range(1, 10)) #avoids 0
problem.addVariables("WOUR", range(10))

#constraints
def myConstraint(t, w, o, u, r, f):
    if 2 * (t * 100 + w * 10 + o) == f * 1000 + o * 100 + u * 10 + r:
        return True
    
problem.addConstraint(myConstraint, "TWOURF")
problem.addConstraint(constraint.AllDifferentConstraint()) #from python library
solutions = problem.getSolutions()
for s in solutions:
    print(s)
```

```python
import constraint
problem = constraint.Problem()

problem.addVariable("p", range(1, 100))
problem.addVariable("r", range(1, 20))
problem.addVariable("s", range(1, 10))

def numConstraint(p, r, s):
    if p + r + s >= 80:
        return True

def priceConstraint(p, r, s):
    if (p*1 + r*5 + s*10) == 100:
        return True

problem.addConstraint(numConstraint, ["p", "r", "s"])
problem.addConstraint(priceConstraint, ["p", "r", "s"])

solutions = problem.getSolutions()

for s in solutions:
    print (s)
```

```python
import constraint
problem = constraint.Problem()
t = int(input("Enter the cost of the drink: "))
p = 0
while ((p<t) or (p>50)):
    p = int(input("Enter cash: "))

problem.addVariable("a", range(6))
problem.addVariable("b", range(5))
problem.addVariable("c", range(3))
problem.addVariable("d", range(3))
problem.addVariable("e", range(2))

def changeConstraint(a, b, c, d, e):
    if (a * 1 + b * 5 + c * 10 + d * 20 + e * 40) == (p - t):
        return True

problem.addConstraint(changeConstraint, ["a", "b", "c", "d", "e"])
solutions = problem.getSolutions()

#optimization - the best solution
coins = 10000
for s in solutions:
    coinCount = s["a"] + s["b"] + s["c"] + s["d"] + s["e"]
    print(s)
    if coinCount < coins:
        coins = coinCount
        bestSolution = s
print("Best solution is", bestSolution)
```

```python
#import the constraint
import constraint
#define the problem
problem = constraint.Problem()

#adding the variables - problem.addVariable
problem.addVariable("WA", ["R", "G", "B"])
problem.addVariable("NT", ["R", "G", "B"])
problem.addVariable("SA", ["R", "G", "B"])
problem.addVariable("QL", ["R", "G", "B"])
problem.addVariable("NSW", ["R", "G", "B"])
problem.addVariable("V", ["R", "G", "B"])
problem.addVariable("T", ["R", "G", "B"])

#define the constraint
def myConstraint(WA, NT, SA, QL, NSW, V, T):
    if (WA != NT) and (WA != SA) and \
    (SA != NT) and (SA != QL) and (SA != NSW) and (SA != V) and \
    (NT != QL) and (NSW != QL) and (NSW != V) and (V != T):
        return True

#use the constraint
problem.addConstraint(myConstraint, ["WA", "NT", "SA", "QL", "NSW", "V", "T"])
solutions = problem.getSolutions()
for s in solutions:
    print(s)
```
